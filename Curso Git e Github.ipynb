# Ciclo de vidas dos arquivos
---


- **Untracked**: é o momento em que o arquivo acabou de ser adicionado no repositório mas ainda nao foi visto pelo Git.  

- **Unmodified**: O arquivo foi visto pelo Git, ou seja, ele existe, porém ainda não teve nenhuma modificação.  

- **Modified**: Se editarmos esse arquivo.  

- **Staged**: Depois que foi modificado ele é jogado numa area que é criada a versão. Depois que a versão for fechada/terminada ela é levada (commit) ai passa a ser unmodified de uma nova versão.

**git status** serve para reportar como esta o repositório agora 

**git add readme.md** ou outro nome de arquivo, adiciona o arquivo ao git (ele passar a ser unmodified). Já podemos criar uma versão dele (ser commitado)

Se modificarmos o arquivo, ele precisa ser commitado novamente com **git add**

O commit é o momento que avisamos o Git pegue os arquivos e crie uma imagem dele (versão)

**git commit -m "add readme.md"** criando um commit, -m serve para avisar que vamos passar uma msg que está entre aspas. Essa msg serve para sabermos o que fizemos na versão, prática saudável. **git cmommit -am "msg " comitta direto**
>[master (root-commit) d123215] add readme.md
 1 file changed, 3 insertions(+)
 create mode 100644 readme.md
 
Se digitarmos **git status** vamos reparar que não tem mais nada para ser feitos, ou seja, as mudanças feitas antes já foram salvas.
>On branch master
nothing to commit, working tree clean




# Visualizando logs
---

**git log** mostra as informações sobre a modificação

**git log --decorate** mostra as informações e de qual branch para qual

**git log --author="jho"**  filtrar por autor os commits

**git shortlog** mostra por ordem alfabética os autores e quantos commits fez

**got shortlog -sn** quantidade de commit e a pessoa que realizou o commit

**git log --graph** 

**git show n°_da_hash** mostra o que foi feito nessa hash(n° da versão)

**git diff** mostra a modificação antes de realizar o commit, após a alteração

**git diff --name-only** diz somente o nome do arquivo que foi modificado

**git checkout readme.md** retorna o arquivo para antes da edição antes de add ao commit

**git reset HEAD readme.md** volta uma classe, ou seja, ele tava no estágio de staged e voltou para modified e com isso podemos fazer **git checkout readme.md**

**git reset --soft or --mixed or --hard + hash_do_commit_anterior_que_queremos_retornar** depois do commit, o soft volta para o estágio de staged, o mixed para antes do staged e o hard ignora tudo o que foi feito

# Clonar e Fork
---

**git clone** clona todo o meu repositório do github para meu repositório local

**git clone chave_SSH nome_do_novo_repositório** = a chave SSH fica em cima do repositório que queremos clonar no github e lembrar que precisamos estar na raíz da onde criamos o novo repositório

O clone é util para trabalhar em vários computadores com o mesmo repositório 

**Fork** = pega um projeto do github que não é meu e faz uma cópia para mim. Caso eu faça modificações eu posso enviar um pull request para a pessoa avisando quais foram 

**branch** = o principal chama *master*. Definido como um ponteiro móvel que leva a um commit. 

    Vantagens: Podemos modificar os arquivos sem alterar o local principal. Multiplas pessoas podem trabalhar simultaneamente. Evita conflitos.
 

# Branchs
---

Criando branch = **git checkout -b nome_do_chanch** ele já entra automaticamente nesse branch

**git branch** mostra quais são os existentes e qual estou no momento

**git checkout nome_do_branch** muda de um branch para o outro

**git branch -D nome_do_branch** paga um branch


**Unindo os branchs** - Merge = operação não destrutiva, cria um novo commit para juntar tudo, porém precisamos de um commit extra para isso deixando o histórico poluído.

**Unindo os branchs** - Rebase = pega o que tem no branch separado e coloca no ínicio da fila do branch master. Não criamos uma árvore, pernacemos de forma linear, porém perdemos a ordem cronológica. 

**git merge nome_do_branch**

**git log --graph** mostra os desenhos dos branchs

**git rebase nome_do_branch**
